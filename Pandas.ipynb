DataFrame 열과 행 바꾸기 (Transpose)

```python
new_col_list = excel_db['구분'] #새로운 열 이름 List 반환
```

```python
excel_db_T = excel_db.T #Transpose
```

```python
excel_db_T.columns = new_col_list #열 이름을 기존 '구분' Column으로 변경
```

```python
excel_db_T.index #excel_db_T에 대하여 index 확인, '구분' index 삭제 필요
```

```python
excel_db_T = excel_db_T.loc[1:,:] 
```

```python
excel_db_T['2013 매출총이익']
excel_db_T['2013 매출총이익'].mean() #transpose해준 excel_db에 대하여 2013 매출총이익 평균 계산
```

DataFrame 그대로 사용 but index 바꾸기

```python
excel_db.loc[9,:].mean() #구분 Column은 문자열(string)이여 숫자 계산 불가
```

```python
excel_db_temp = excel_db.copy() #excel_db_temp는 excel_db 원본에서 복사하기
```

```python
excel_db_temp.index = excel_db_temp['구분'] #구분 column을 index로 사용
del excel_db_temp['구분'] #구분 열 제거 (숫자만 남길 수 있도록)
```

```python
excel_db_temp.loc['2013 매출총이익'].mean() #transpose 해준 경우와 동일한 결과 확인 가능
```
