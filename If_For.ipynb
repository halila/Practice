```python
import pandas as pd

Stock_db = pd.DataFrame()
ticker_list = ['AAPL','AMZN','GOOG','META','TSLA']

for ticker in ticker_list:
    Stock_db_temp = pd.read_csv(ticker + '.csv')
    Stock_db_temp['Ticker'] = ticker
    Stock_db = pd.concat([Stock_db, Stock_db_temp])
```

```python
Stock_db.groupby('Ticker').max()
```
